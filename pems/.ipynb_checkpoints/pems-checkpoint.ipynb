# Read csv with full speed

```python
data = pd.read_csv("./data/2017_03_01_full_speed.csv", index_col = 0)
```

```python
def z_score(x, mean, std):
    '''
    Z-score normalization function: $z = (X - \mu) / \sigma $,
    where z is the z-score, X is the value of the element,
    $\mu$ is the population mean, and $\sigma$ is the standard deviation.
    :param x: np.ndarray, input array to be normalized.
    :param mean: float, the value of mean.
    :param std: float, the value of standard deviation.
    :return: np.ndarray, z-score normalized array.
    '''
    return (x - mean) / std
```

```python
data.head()
```

```python
speed = data['Avg Speed full']
```

```python
speed.head(10)
```

```python
speed_mean = np.mean(speed.values)
```

```python
speed_std = np.std(speed.values)
```

```python
speed_mean
```

```python
speed_std
```

```python
speed = z_score(speed, speed_mean, speed_std)
```

```python
speed.head(10)
```

# unstack

```python
data = pd.read_csv("./data/2017_03_01_full_speed.csv", index_col = 0)
```

```python
data.head()
```

```python
new_data = data[['Timestamp','Station','Avg Speed full']]
```

```python
speed_index = new_data.set_index(["Timestamp", "Station"])["Avg Speed full"]
```

```python
speed_index
```

```python
new_df = speed_index.unstack()
```

```python
new_df
```

```python
new_df = new_df.rename_axis(columns=None)
```

```python
new_df.head()
```

```python
new_df.to_csv("./data/2017_03_01_full_speed_inline.csv")
```
